```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('african-regional-energy-statistics-2014.csv')# read the csv file
```

```python
data
```

```python
data['RegionName'].unique()# show the unique values of RegionName column
```

```python
# filter rows according to a RegionName
egdata=data[(data.RegionName=='Egypt')]#|(data.RegionName=='Algeria')|(data.RegionName=='Southern Africa')]
algdata=data[(data.RegionName=='Algeria')]
algdata
```

```python
f= egdata['IndicatorName']== 'Final Consumption of coking coal, 1000 tonnes'# filter the indicator

egdata=egdata[f]
```

```python
f1= algdata['IndicatorName']== 'Final Consumption of coking coal, 1000 tonnes'# filter the indicator

algdata=algdata[f1]
```

```python
egdata=egdata[['Date','Value']]# having only the Date and Value columns
egdata=egdata.rename({'Value':'Egypt'}, axis=1)# rename the column Value to be Egypt
```

```python
algdata=algdata[['Date','Value']]# having only the Date and Value columns
algdata
```

```python
algdata=algdata.rename({'Value':'Algeria'}, axis=1)
```

```python
egdata=egdata.sort_values(by='Date',ascending=True)# sort the dataFrame
algdata=algdata.sort_values(by='Date',ascending=True)# sort the dataFrame
```

```python
Data=pd.merge(egdata,algdata, on="Date") # merge the two dataframes 
```

```python
# visualize the data
plt.figure(figsize=(8,5)) #adjusting the figure size

plt.title('Consumption of coking coal in Egypt', fontdict={'fontweight':'bold', 'fontsize': 12})# title
plt.plot(Data.Date, Data.Egypt, 'b.-', label='Egypt')# 
plt.plot(Data.Date, Data.Algeria, 'r.-',  label='Algeria')

plt.xlabel('Year')
plt.ylabel('Consumption')
plt.legend()
#plt.savefig('Gas_price_figure.png', dpi=300)
plt.show()
```
